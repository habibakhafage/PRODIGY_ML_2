```python
from scipy import stats

z_scores = stats.zscore(df['Annual Income (k$)'])
threshold = 3
outliers = (np.abs(z_scores) > threshold)
df_no_outliers = df[~outliers]
```

```python
Q1 = df['Annual Income (k$)'].quantile(0.25)
Q3 = df['Annual Income (k$)'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['Annual Income (k$)'] >= lower_bound) & (df['Annual Income (k$)'] <= upper_bound)]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_no_outliers['Annual Income (k$)'])
plt.title('Annual Income (k$) Boxplot')
plt.show()
```

```python
outliers_mask = (df['Annual Income (k$)'] < lower_bound) | (df['Annual Income (k$)'] > upper_bound)
clean_data = df[~outliers_mask]
num_outliers_removed = len(df) - len(df_no_outliers)
print(f'Number of outliers removed: {num_outliers_removed}')
```

```python
for col in df_no_outliers.describe().columns:
    plt.figure(figsize=(16,2))
    sns.boxplot(data=df_no_outliers, x=col)
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
data_to_normalize = df_no_outliers[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(data_to_normalize)
normalized_MinMax_df = pd.DataFrame(normalized_data, columns=data_to_normalize.columns)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
finaldf = pd.DataFrame(sc.fit_transform(normalized_MinMax_df),columns=normalized_MinMax_df.columns)
finaldf
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(finaldf)
finaldf['PCA1'] = pca_components[:, 0]
finaldf['PCA2'] = pca_components [:, 1]
```

```python
finaldf
```

```python
df2 = finaldf[['PCA1','PCA2']]
df2
```

# 5 clusters 

```python
df2.head()
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(df2)
    wcss.append(km.inertia_)
```

```python
#The elbow curve
plt.figure(figsize=(12,6))
plt.plot(range(1,11),wcss)
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
km1=KMeans(n_clusters=5,random_state=42)
km1.fit(df2)
```

```python
y=km1.predict(df2)
```

```python
df2["label"] = y
df2.head()
```

```python
df2['label'].unique()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'PCA1',y = 'PCA2',hue="label",  
                 palette=['green','orange','brown','dodgerblue','red'], legend='full',data = df2  ,s = 60 )
plt.xlabel('PCA1')
plt.ylabel('PCA2') 
plt.title('PCA2 vs PCA2')
plt.show()
```

```python
# K-Means algorithm generated the following 5 clusters:

#  clients with low annual income and low spending score
#  clients with medium annual income and low spending score
#  clients with high annual income and low spending score
#  clients with low annual income and high spending score
#  clients with high annual income and high spending score
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(df2,y)
print(f'silhouette score for 5 clusters: {score:.4f}')
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans(random_state=22)
visualizer = KElbowVisualizer(model, k=(2,10), metric = 'silhouette')
visualizer.fit(df2)
visualizer.show()
plt.show()
```

```python
# Silhouette score method indicates the best options would be 4 or 5 or 6 clusters. 
```
